```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we generate data of the form y = ax + e (where e is Gaussian noise)
a = 0.35

nr_datapoints = 500
x_data = np.array(range(0, nr_datapoints+1))/nr_datapoints
y_data = a*x_data + np.random.normal(0,1,nr_datapoints+1)
y_val = a*x_data
```

```python
# next we create a neural net in Keras that is going to perform simple linear regression, and should find the value of a
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(1, activation=None, input_shape=(1,)))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
# Generate dummy data
import numpy as np
data = x_data
labels = y_data

model.fit(
    data, 
    labels, 
    batch_size=1, # how many example do we train in tandem?
    epochs=10, # how may times do we go through the entire dataset
    verbose=0, # set to '2' to si printing the progress bar
    callbacks=None, 
    validation_split=0.2, 
    validation_data=None, 
    shuffle=True, #shuffle the data
    class_weight=None, 
    sample_weight=None, 
    initial_epoch=0)
```

```python
test_data = np.array(range(0, nr_datapoints+1))/nr_datapoints

model_predictions = model.predict(
    test_data, 
    batch_size=1, 
    verbose=1)
```

```python
plt.figure(figsize=(10,4))
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x_data, y_data)
plt.plot(x_data, y_val, color='red')
plt.scatter(test_data, model_predictions)
plt.legend(['linear model', 'data', 'predictions'])
plt.show()
```
